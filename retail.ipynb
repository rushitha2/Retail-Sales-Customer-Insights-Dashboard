### Connecting database

### customer data

```python
df = pd.read_json('customers.json')
df
```

```python
df.isnull().sum()
```

```python
c= df.drop_duplicates()
c
```

```python
c.isnull().sum()
```

#### handling gender and region column values

```python
uniq= c['Gender'].unique()
print(uniq)
```

```python
c.loc[:,'Gender'] = c['Gender'].str.lower()
```

```python
gender = {'m': 'male', 'male': 'male', 'f': 'female','female': 'female'}

c.loc[:,'Gender'] = c['Gender'].map(gender)
```

```python
uniq= c['Gender'].unique()
print(uniq)
```

```python
uniq= c['Region'].unique()
print(uniq)
```

```python
c.loc[:,'Region'] = c['Region'].str.lower()
```

```python
region= {'ohho':'ohio','ohio':'ohio', 'california':'california', 'californiya':'california','new yorkk':'new york',
         'new york':'new york','ny':'new york','nw york':'new york','texas': 'texas', 'texaz':'texas'}

c.loc[:,'Region'] = c['Region'].map(region)
```

```python
c.loc[:,'Region'] = c['Region'].fillna('Unknown')
```

```python
uniq= c['Region'].unique()
print(uniq)
```

```python
c
```

#### inserting into customer table

```python
to_insert = [tuple(row) for row in c.to_numpy()]

insert_query = 'INSERT INTO dim_customer (customerid,firstname,lastname,gender,region,ssn) VALUES (%s, %s,%s,%s,%s,%s)'

cur.executemany(insert_query, to_insert)
db.commit()
```

### Sale data

```python
df = pd.read_csv('sales 1.csv')
df
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

#### handling null values

```python
df.dropna(subset=['SalesAmount', 'Quantity'], how='all', inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
x= df['SalesAmount'].mean()
x
```

```python
msa = df['SalesAmount'].median()
msa
```

```python
df.loc[:, 'SalesAmount'] = df['SalesAmount'].fillna(msa)

df.isnull().sum()
```

```python
mq = int(df['Quantity'].median())
mq
```

```python
df.loc[:,'Quantity']= df['Quantity'].fillna(mq)

df.isnull().sum()
```

```python
df
```

#### importing sale data into mysql

```python
to_insert = [tuple(row) for row in df.to_numpy()]

insert_query = 'INSERT INTO fact_sales (saleid, productid,customerid, salesamount, quantity,timestamp)VALUES (%s, %s,%s,%s,%s,%s)'

cur.executemany(insert_query, to_insert)
db.commit()
```
